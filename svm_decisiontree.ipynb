```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("iris.csv")
data
```

```python
x=data[["sepal.length","sepal.width","petal.length","petal.width"]]
y=data["variety"]
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
p=log.predict(x_test)
p
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix
a=accuracy_score(y_test,p)
a
```

```python
pr=confusion_matrix(y_test,p)
pr
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
predict_knn=knn.predict(x_test)
predict_knn
```

```python
b=accuracy_score(y_test,predict_knn)
b
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(a,b,'o')
plt.show()
```

```python
plt.scatter(a,b,color='red')
plt.show()
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear')  #kernal can be linear,#poly,#sigmoid #rbf
```

```python
s=svm.fit(x_train,y_train)
```

```python
svm_predict=svm.predict(x_test)
acc_svm=accuracy_score(y_test,svm_predict)
acc_svm
```

```python
svm_predict=svm.predict(x_test)
```

```python
svm_predict
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
dt_pr=clf.predict(x_test)
acc_dt=accuracy_score(y_test,dt_pr)
acc_dt
```

```python
plt.scatter(acc_svm,acc_dt,color='red')
plt.show()
```

```python
plt.bar(a,b,acc_svm,acc_dt,color='red')
plt.show()
```
